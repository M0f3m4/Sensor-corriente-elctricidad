### <span style="color:blue">Definición de Problema </span>

<b> Sistema de Monitoreo de Energía Doméstica: </b>

El problema del consumo de energía doméstica radica en el uso excesivo e ineficiente de recursos energéticos en nuestros hogares, lo que tiene múltiples consecuencias negativas. Este consumo desmedido contribuye al agotamiento de recursos naturales y al aumento de emisiones de gases de efecto invernadero, exacerbando el cambio climático. Además, el alto consumo energético en los hogares conlleva costos económicos significativos para las familias y puede provocar apagones y sobrecargas en las redes eléctricas.

Para resolver este problema, crearemos un sistema que rastree el consumo de energía en el hogar y ofrezca recomendaciones para reducir el consumo y ahorrar dinero.

### <span style="color:blue"> Marco Teórico </span>

“El sistema de monitoreo te permite visualizar y analizar en tiempo real el consumo de energía de tu hogar o negocio para que puedas ver todos los detalles sobre el consumo total, la demanda de energía en diferentes momentos del día, los patrones de uso y otros datos que, sin duda, son muy relevantes.
Algunos sistemas de monitoreo también ofrecen funciones adicionales, como la capacidad de establecer alarmas o notificaciones para alertarle sobre un consumo anormalmente alto, la posibilidad de establecer metas de ahorro de energía y recibir recomendaciones personalizadas para optimizar tu consumo.” (Almenara, 2023)


“Del análisis de la encuesta se determinó que la mayoría de las familias encuestadas no toman conciencia del consumo eléctrico en sus hogares, originando elevados costos en los recibos, sobrecarga eléctrica en algunos electrodomésticos, ocasionando su deterioro y su contaminación ambiental, entre otros.” (Montes, 2023)



### <span style="color:blue"> Protocolo Sensor de Corriente </span>
Para crear un sistema de monitoreo de energía doméstica utilizando un sensor de corriente y una placa STM32F767ZIT6, podemos utilizar varios protocolos de comunicación, pero el protocolo  más adecuado para esta aplicación es el protocolo UART (Universal Asynchronous Receiver/Transmitter).

 **UART (Universal Asynchronous Receiver/Transmitter):** UART es un protocolo de comunicación serie simple y ampliamente utilizado. Requiere dos pines, uno para transmitir (TX) y otro para recibir (RX). Puedes conectar el sensor de corriente al microcontrolador a través de UART y programar el STM32 para leer y procesar los datos del sensor de corriente. Este enfoque es adecuado si solo necesitas la comunicación punto a punto entre el sensor y el microcontrolador.

**SPI (Serial Peripheral Interface):** SPI es otro protocolo de comunicación serie que permite una comunicación más rápida y es especialmente útil cuando necesitas comunicarte con múltiples dispositivos en un bus SPI. Si planeas agregar más sensores o dispositivos en el futuro, el uso de SPI podría ser una opción adecuada. Sin embargo, SPI requiere más pines que UART, generalmente al menos cuatro pines: SCLK (reloj de serie), MISO (maestro hacia esclavo), MOSI (esclavo hacia maestro) y SS/CS (selección de esclavo/chip). Asegúrate de que la placa STM32F767ZIT6 tenga suficientes pines SPI disponibles.

### <span style="color:blue"> Descripción de Solución </span>
<b>1. Sección de Inicialización: </b>

El código comienza con información del encabezado e incluye bibliotecas y definiciones necesarias.
Inicializa periféricos como el ADC (Convertidor Analógico a Digital) y el UART (Receptor-Transmisor Universal Asíncrono).

<b> 2. Función Principal: </b>

La función main es el punto de entrada del programa.
Ingresa a un bucle infinito, monitoreando continuamente la corriente eléctrica.

<b> 3. Configuración del ADC: </b>

El ADC (Convertidor Analógico a Digital) se configura para leer señales analógicas.
El canal 9 del ADC se configura para muestrear la corriente eléctrica, como indica sConfig.Channel = ADC_CHANNEL_9.

<b>4. Configuración del UART: </b>

El UART (Receptor-Transmisor Universal Asíncrono) se configura para transmitir datos.
Se utiliza UART3 con una velocidad de baudios de 115200, 8 bits de datos, 1 bit de parada y sin paridad.

<b> 5. Bucle Principal: </b>

Dentro del bucle infinito, el código realiza las siguientes acciones repetidamente:
Inicia la conversión ADC (HAL_ADC_Start).
Espera a que la conversión se complete (HAL_ADC_PollForConversion).
Obtiene el resultado de la conversión ADC (HAL_ADC_GetValue).
Convierte el valor ADC crudo a una cadena (sprintf).
Transmite la cadena a través del UART (HAL_UART_Transmit).
Espera un segundo (HAL_Delay(1000)).

<b> 6. Configuración del Reloj del Sistema: </b>

El reloj del sistema se configura para utilizar el oscilador interno de alta velocidad (HSI) a una frecuencia de aproximadamente 16 MHz.

<b> 7. Inicialización de GPIO: </b>

Los puertos GPIO B y D se habilitan, en los pines PD8, PD9, USART3_Rx y USART3_TX.

En resumen, el código monitorea continuamente la corriente eléctrica utilizando un ADC, convierte los valores ADC crudos a cadenas y los transmite a través de UART. Los pines GPIO específicos utilizados para el monitoreo de corriente y otras funcionalidades pueden estar configurados en otras partes del código que no están incluidas en el fragmento.
